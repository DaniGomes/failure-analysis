```python
def draw_rules(rules, confidence):
    rules = rules[rules['confidence']>confidence]
    G1 = nx.DiGraph()
    mapa_de_cores=[]
    N = 2000
    cores = np.random.rand(N) 
      
    for i in range (len(rules)):
     
        for c in rules.iloc[i]['consequents']:
            G1.add_nodes_from([c])
        
        for a in rules.iloc[i]['antecedents']:
            G1.add_nodes_from([a], fillcolor='pink')
            G1.add_edge(a, c, color=cores[i] , weight = 2)
            
    for node in G1:
        if(node.endswith('_R') or node.endswith('-R')):
            mapa_de_cores.append('red')
        else:
            mapa_de_cores.append('pink')
            
    edges = G1.edges()
    cores = [G1[u][v]['color'] for u,v in edges]
    weights = [G1[u][v]['weight'] for u,v in edges]
 
    pos = nx.spring_layout(G1, k=16, scale=1)
    nx.draw_networkx(G1, pos, edges=edges, node_color=mapa_de_cores, edge_color=cores, width=weights, font_size=16, with_labels=False)            
   
    for p in pos:
        pos[p][1] += 0.07
    nx.draw_networkx_labels(G1, pos)
    plt.show()
```

```python
plt.figure(figsize=(15,6))
print("Please choose a dataset: ")
data = widgets.Dropdown(
    options=['unclustered', 'clustered', 'INF', 'CS'],
    value='unclustered',
    description='Dataset:',
    disabled=False,
)
display(data)

print("Confidence over:")
conf = widgets.IntSlider(
    value=70,
    min=70,
    max=100,
    step=1,
    description='Value:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
display(conf)

button = widgets.Button(description="Visualize")
output = widgets.Output()

display(button, output)

def on_button_clicked(b):
    print(data.value, ' with confidence at least ', conf.value, '%')
    if data.value == 'unclustered':
        draw_rules(unclus, conf.value/100)
    elif data.value == 'clustered':
        draw_rules(clus, conf.value/100)
    elif data.value == 'INF':
        draw_rules(inf, conf.value/100)
    elif data.value == 'CS':
        draw_rules(ccp, conf.value/100)
    
button.on_click(on_button_clicked)
```
